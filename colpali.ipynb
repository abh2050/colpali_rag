```python
import base64
import os
from byaldi import RAGMultiModalModel
```

```python
RAG = RAGMultiModalModel.from_pretrained('vidore/colpali-v1.2',verbose=1)
```

```python
!pip install gdown # Install the gdown library to download files from Google Drive

import gdown # Import the gdown library

# Extract the file ID from the URL
url = 'https://drive.google.com/file/d/1UzEbtQTleX82vd7WwdqRxeXrPueSCdlr/view?usp=drive_link'
file_id = url.split('/')[-2]

# Create the direct download link using the file ID
download_url = f'https://drive.google.com/uc?id={file_id}'

# Download the file
output = 'doc.pdf'  # Set the desired output file name
gdown.download(download_url, output, quiet=False)
```

```python
RAG.index(
    input_path= "/content/doc.pdf",
    index_name='colpali',
    store_collection_with_index=True,
    overwrite= True
)
```

```python
query = 'list the parts for the master brake cylinder and pin installation in a table'
```

```python
results = RAG.search(query,k=1)
```

```python
results
```

```python
image_bytes = base64.b64decode(results[0].base64)
```

```python
filename = 'image.png'
with open(filename, 'wb') as f:
    f.write(image_bytes)
```
